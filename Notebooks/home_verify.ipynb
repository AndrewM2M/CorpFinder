```python
import extruct
from w3lib.html import get_base_url
import itertools
from extruct.jsonld import JsonLdExtractor
from fuzzywuzzy import fuzz
```

```python
sources = dict(zip(data_set["Company Name"], data_set["URL"]))
for company, site in itertools.islice(sources.items(), 0, 10):
    print(f"{company}, {site}")
```

```python
jslde = JsonLdExtractor()
jld_data = {}
def json_ld_check(sources):
    jld_data = {}
    for company, site in itertools.islice(sources.items(), 0, 10):
        print (f"checking: {company} -> {site}")
        try:
            html = url_fetch.get(site)
        except:
            print(f"{company} got a url error")
            jld_data[company] = "Url Error"
            continue
        base_url = get_base_url(html.text, html.url)
        jLD = jslde.extract(html.text)
        for i in jLD:
            found_name = ""
            if "name" in i.keys():
                found_name = i["name"]
                ratio = fuzz.token_set_ratio(i["name"], company)
                good = True if ratio > 96 else False
            if good and site not in jld_data:
                print(f"{found_name} maches {company} @ {ratio}")
                jld_data[company] = "Url Good"
                break
            else:
                print(f"{found_name} maches {company} @ {ratio}")
                jld_data[company] = "Check Url"
    return jld_data
```

```python
jld_data=json_ld_check(sources)
```
