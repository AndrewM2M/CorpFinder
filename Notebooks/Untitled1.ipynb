```python
import pprint
import os
import pandas as pd
from docx import Document


mydir = r"C:\Users\andrew.white\Documents\Lists\WD Numeric"
docs = os.listdir(mydir)
Sources = set()
records = []
columns = {"Source"}
rec_dict = {"Source": []}
idx = 0

for file in docs:
    bold_list = set()
    record_values = set()
    if file.endswith(".docx"):
        docx = Document(os.path.join(mydir, file))
        Sources.add(file)
        idx = abs(idx) + 1
    else:
        continue
    for pidx, para in enumerate(docx.paragraphs, 1):
        if para.text.find(": "):
            for run in para.runs:
                if run.bold:
                    try:
                        rec_idx
                    except NameError:
                        rec_idx = idx+(pidx*0.1)
                        records.append((rec_idx, "Source", file))
        c = para.text.split(": ", 1)[0].strip()
        c = c.rstrip(":.?,")
        try:
            v = para.text.split(": ", 1)[1].strip()
        except IndexError:
            v = None
        columns.add(c)
        record_values.add(v)
        records.append((rec_idx, c, v))
```

```python
df = pd.DataFrame.from_records(records)
```
