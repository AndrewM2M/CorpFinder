```python
import pywikibot
from pywikibot.data import api
from pywikibot import pagegenerators
from fuzzywuzzy import fuzz
from fuzzywuzzy import process
```

```python
site = pywikibot.Site()
```

```python
cat = pywikibot.Category(site,'Category:Living people')
```

```python
gen = pagegenerators.CategorizedPageGenerator(cat)
```

```python
def get_items(site, item_title):
    """
    Requires a site and search term (item_title) and returns the results.
    """
    params = {"action": "wbsearchentities",
              "format": "json",
              "language": "en",
              "type": "item",
              "search": item_title}
    request = api.Request(site=site, parameters=params)
    return request.submit()
```

```python
hits = get_items(site, "Intel")
```

```python
hit_keys = hits.keys()
```

```python
goodies = hits.get("search")
```

```python
ids = set()
for e in goodies:
    ids.add((e.get("label"), e.get("id")))
```

```python
ids
```

```python
token_set = process.extract("intel", ids, limit=3,scorer=fuzz.token_set_ratio)
token_sort = process.extract("intel", ids, limit=3,scorer=fuzz.token_sort_ratio)
set_ratio = sum(x[1] for x in token_set)
sort_ratio = sum(x[1] for x in token_sort)
```

```python
final_ids = token_set if set_ratio > sort_ratio else token_sort
```

```python
final_ids
```

```python
repo = site.data_repository()
token = repo.token(pywikibot.Page(repo, 'Main Page'), 'edit')
```

```python
def getItem(site, wdItem, token):
    request = api.Request(site=site,
                          action='wbgetentities',
                          format='json',
                          ids=wdItem)    
    return request.submit()
```

```python
data = []
for l in final_ids:
    data.append(getItem(site,l[0][1],token))
```
